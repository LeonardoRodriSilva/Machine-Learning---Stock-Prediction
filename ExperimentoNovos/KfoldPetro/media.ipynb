```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do CSV
data = pd.read_csv("..\PadraoPetro\model_performance_all_days.csv")  # Substitua "caminho_do_arquivo.csv" pelo caminho real do seu arquivo

# Configurar o tamanho da figura
plt.figure(figsize=(14, 10))

# Ajustar os parâmetros de fonte
plt.rcParams.update({'font.size': 14})

# Subplot 1 - MAE ao longo dos dias
plt.subplot(2, 2, 1)
plt.plot(data['Days'], data['MAE_Train'], marker='o', label='Treino')
plt.plot(data['Days'], data['MAE_Test'], marker='o', label='Teste')
plt.title('MAE ao longo dos dias')
plt.xlabel('Dias')
plt.ylabel('MAE')
plt.xticks(range(1, 21))
plt.legend()

# Subplot 2 - MSE ao longo dos dias
plt.subplot(2, 2, 2)
plt.plot(data['Days'], data['MSE_Train'], marker='o', label='Treino')
plt.plot(data['Days'], data['MSE_Test'], marker='o', label='Teste')
plt.title('MSE ao longo dos dias')
plt.xlabel('Dias')
plt.ylabel('MSE')
plt.xticks(range(1, 21))
plt.legend()

# Subplot 3 - RMSE ao longo dos dias
plt.subplot(2, 2, 3)
plt.plot(data['Days'], data['RMSE_Train'], marker='o', label='Treino')
plt.plot(data['Days'], data['RMSE_Test'], marker='o', label='Teste')
plt.title('RMSE ao longo dos dias')
plt.xlabel('Dias')
plt.ylabel('RMSE')
plt.xticks(range(1, 21))
plt.legend()

# Subplot 4 - R² ao longo dos dias
plt.subplot(2, 2, 4)
plt.plot(data['Days'], data['R²_Train'], marker='o', label='Treino')
plt.plot(data['Days'], data['R²_Test'], marker='o', label='Teste')
plt.title('R² ao longo dos dias')
plt.xlabel('Dias')
plt.ylabel('R²')
plt.xticks(range(1, 21))
plt.legend()

# Ajustar o espaçamento entre os subplots e exibir o gráfico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy as np

media_desvio_fold = []

dir_name = "KfoldPetro"
filename = os.path.join(dir_name, "model_performance_all_kfold_avg.csv")

data_kfold = pd.read_csv(filename)

for days in range(1, 21):
    fold_data = data_kfold[data_kfold['Days'] == days]
    
    if not fold_data.empty:
        media_mae_kfold = np.mean(fold_data['Avg MAE'])
        desvio_mae_kfold = np.std(fold_data['Avg MAE'])
        media_mse_kfold = np.mean(fold_data['Avg MSE'])
        desvio_mse_kfold = np.std(fold_data['Avg MSE'])
        media_rmse_kfold = np.mean(fold_data['Avg RMSE'])
        desvio_rmse_kfold = np.std(fold_data['Avg RMSE'])
        media_r2_kfold = np.mean(fold_data['Avg R²'])
        desvio_r2_kfold = np.std(fold_data['Avg R²'])

        media_desvio_fold.append([days, media_mae_kfold, desvio_mae_kfold, 
                                   media_mse_kfold, desvio_mse_kfold, 
                                   media_rmse_kfold, desvio_rmse_kfold, 
                                   media_r2_kfold, desvio_r2_kfold])

df_resultados_fold = pd.DataFrame(media_desvio_fold, columns=['Days', 'MAE_Mean', 'MAE_Std', 
                                                               'MSE_Mean', 'MSE_Std', 
                                                               'RMSE_Mean', 'RMSE_Std', 
                                                               'R²_Mean', 'R²_Std'])

df_resultados_fold.to_csv("media_desvio_resultados_kfold_folds.csv", index=False)

plt.figure(figsize=(14, 10))


plt.subplot(2, 2, 1)
plt.errorbar(df_resultados_fold['Days'], df_resultados_fold['MAE_Mean'], 
             yerr=df_resultados_fold['MAE_Std'], marker='o', capsize=5, color='blue')
plt.title('MAE ao longo do tempo PETRO')
plt.xlabel('Days')
plt.ylabel('MAE_Mean')
plt.xticks(np.arange(1, 21, 1))  
plt.legend()


plt.subplot(2, 2, 2)
plt.errorbar(df_resultados_fold['Days'], df_resultados_fold['MSE_Mean'], 
             yerr=df_resultados_fold['MSE_Std'], marker='o', capsize=5, color='orange')
plt.title('MSE ao longo do tempo PETRO')
plt.xlabel('Days')
plt.ylabel('MSE_Mean')
plt.xticks(np.arange(1, 21, 1))  
plt.legend()

plt.subplot(2, 2, 3)
plt.errorbar(df_resultados_fold['Days'], df_resultados_fold['RMSE_Mean'], 
             yerr=df_resultados_fold['RMSE_Std'], marker='o', capsize=5, color='green')
plt.title('RMSE ao longo do tempo PETRO')
plt.xlabel('Days')
plt.ylabel('RMSE_Mean')
plt.xticks(np.arange(1, 21, 1))  
plt.legend()

plt.subplot(2, 2, 4)
plt.errorbar(df_resultados_fold['Days'], df_resultados_fold['R²_Mean'], 
             yerr=df_resultados_fold['R²_Std'], marker='o', capsize=5, color='red')
plt.title('R² ao longo do tempo PETRO')
plt.xlabel('Days')
plt.ylabel('R²_Mean')
plt.xticks(np.arange(1, 21, 1))  
plt.legend()

plt.tight_layout()
plt.show()
```

Gráfico petro kfolds

```python
melhor_janela_mae_fold = df_resultados_fold.loc[df_resultados_fold['MAE_Mean'].idxmin()]

melhor_janela_mse_fold = df_resultados_fold.loc[df_resultados_fold['MSE_Mean'].idxmin()]

melhor_janela_rmse_fold = df_resultados_fold.loc[df_resultados_fold['RMSE_Mean'].idxmin()]

melhor_janela_r2_fold = df_resultados_fold.loc[df_resultados_fold['R²_Mean'].idxmax()]

print("Melhor janela baseada no MAE:", melhor_janela_mae_fold)
print("Melhor janela baseada no MSE:", melhor_janela_mse_fold)
print("Melhor janela baseada no RMSE:", melhor_janela_rmse_fold)
print("Melhor janela baseada no R²:", melhor_janela_r2_fold)
```

```python
import matplotlib.pyplot as plt
import os
import pandas as pd

media_desvio_fold_diff = []

for days in range(1, 21):
    dir_name = f"Janela_itau_kfold_diff_{days}"
    filename = os.path.join(dir_name, f"model_performance_{days}_days_kfold_diff.csv")
    
    data_kfold_diff = pd.read_csv(filename)
    
    for fold in range(1, 6):
        fold_data_diff = data_kfold_diff[data_kfold_diff['Fold'] == fold]
        media_mae_kfold_diff = fold_data_diff['MAE'].mean()
        desvio_mae_kfold_diff = fold_data_diff['MAE'].std()
        media_mse_kfold_diff = fold_data_diff['MSE'].mean()
        desvio_mse_kfold_diff = fold_data_diff['MSE'].std()
        media_rmse_kfold_diff = fold_data_diff['RMSE'].mean()
        desvio_rmse_kfold_diff = fold_data_diff['RMSE'].std()
        media_r2_kfold_diff = fold_data_diff['R²'].mean()
        desvio_r2_kfold_diff = fold_data_diff['R²'].std()
        
        media_desvio_fold_diff.append([fold, days, media_mae_kfold_diff, desvio_mae_kfold_diff, media_mse_kfold_diff, desvio_mse_kfold_diff, media_rmse_kfold_diff, desvio_rmse_kfold_diff, media_r2_kfold_diff, desvio_r2_kfold_diff])

df_resultados_fold_diff = pd.DataFrame(media_desvio_fold_diff, columns=['Fold', 'Days', 'MAE_Mean', 'MAE_Std', 'MSE_Mean', 'MSE_Std', 'RMSE_Mean', 'RMSE_Std', 'R²_Mean', 'R²_Std'])

df_resultados_fold_diff.to_csv("media_desvio_resultados_kfold_folds_diff.csv", index=False)

plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
for fold in range(1, 6):
    fold_data_diff = df_resultados_fold_diff[df_resultados_fold_diff['Fold'] == fold]
    plt.errorbar(fold_data_diff['Days'], fold_data_diff['MAE_Mean'], yerr=fold_data_diff['MAE_Std'], marker='o', capsize=5, label=f'Fold {fold}')
plt.title('MAE ao longo do tempo por fold')
plt.xlabel('Days')
plt.ylabel('MAE_Mean')
plt.legend()

plt.subplot(2, 2, 2)
for fold in range(1, 6):
    fold_data_diff = df_resultados_fold_diff[df_resultados_fold_diff['Fold'] == fold]
    plt.errorbar(fold_data_diff['Days'], fold_data_diff['MSE_Mean'], yerr=fold_data_diff['MSE_Std'], marker='o', capsize=5, label=f'Fold {fold}')
plt.title('MSE ao longo do tempo por fold')
plt.xlabel('Days')
plt.ylabel('MSE_Mean')
plt.legend()

plt.subplot(2, 2, 3)
for fold in range(1, 6):
    fold_data_diff = df_resultados_fold_diff[df_resultados_fold_diff['Fold'] == fold]
    plt.errorbar(fold_data_diff['Days'], fold_data_diff['RMSE_Mean'], yerr=fold_data_diff['RMSE_Std'], marker='o', capsize=5, label=f'Fold {fold}')
plt.title('RMSE ao longo do tempo por fold')
plt.xlabel('Days')
plt.ylabel('RMSE_Mean')
plt.legend()

plt.subplot(2, 2, 4)
for fold in range(1, 6):
    fold_data_diff = df_resultados_fold_diff[df_resultados_fold_diff['Fold'] == fold]
    plt.errorbar(fold_data_diff['Days'], fold_data_diff['R²_Mean'], yerr=fold_data_diff['R²_Std'], marker='o', capsize=5, label=f'Fold {fold}')
plt.title('R² ao longo do tempo por fold')
plt.xlabel('Days')
plt.ylabel('R²_Mean')
plt.legend()

plt.tight_layout()
plt.show()
```
