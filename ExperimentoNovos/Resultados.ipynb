```python
# Create lists to store all results
experiments = [
    'Kfold', 
    'KfoldTuner', 
    'KfoldTunerIndicatorV1', 
    'KfoldTunerIndicatorV2',
    'KfoldTunerIndicatorV1Macro',
    'KfoldTunerIndicatorV1Micro',
    'KfoldTunerIndicatorMacro12',
    'KfoldTuner12'
]

data = []
for exp in experiments:
    try:
        results = eval(f'Resultados{exp}')
        
        # Debug print
        print(f"\nChecking {exp}:")
        print(f"Keys available: {results.keys()}")
        
        # Add data with error checking
        data.append({
            'Experiment': exp,
            'MAE': results.get('best_MAE', {}).get('value', float('nan')),
            'MAE Days': results.get('best_MAE', {}).get('days', float('nan')),
            'RMSE': results.get('best_RMSE', {}).get('value', float('nan')),
            'RMSE Days': results.get('best_RMSE', {}).get('days', float('nan')),
            'MSE': results.get('best_MSE', {}).get('value', float('nan')),
            'MSE Days': results.get('best_MSE', {}).get('days', float('nan')),
            'R²': results.get('best_R2', {}).get('value', float('nan')),
            'R² Days': results.get('best_R2', {}).get('days', float('nan'))
        })
    except NameError:
        print(f"Warning: Results for {exp} not found")
        data.append({
            'Experiment': exp,
            'MAE': float('nan'),
            'MAE Days': float('nan'),
            'RMSE': float('nan'),
            'RMSE Days': float('nan'),
            'MSE': float('nan'),
            'MSE Days': float('nan'),
            'R²': float('nan'),
            'R² Days': float('nan')
        })

# Create and style DataFrame
comparison_df = pd.DataFrame(data)
comparison_df.set_index('Experiment', inplace=True)

styled_df = (comparison_df
    .style
    .format({
        'MAE': '{:.4f}',
        'RMSE': '{:.4f}',
        'MSE': '{:.4f}',
        'R²': '{:.4f}'
    })
    .background_gradient(
        cmap='RdYlGn_r',
        subset=['MAE', 'RMSE', 'MSE']
    )
    .background_gradient(
        cmap='RdYlGn',
        subset=['R²']
    )
    .set_caption("Comparison of Best Results Across Experiments"))

display(styled_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\\Users\\leona\\pyhtonscripts\\ExperimentosTeste\\Micro\\metricas_lstm.csv')

def encontrar_melhor_janela(df):
    melhores_janelas = []
    tickers = df['Ticker'].unique()
    
    for ticker in tickers:
        df_ticker = df[df['Ticker'] == ticker]
        melhor_mae = df_ticker.loc[df_ticker['MAE'].idxmin()]
        melhor_rmse = df_ticker.loc[df_ticker['RMSE'].idxmin()]
        melhor_mse = df_ticker.loc[df_ticker['MSE'].idxmin()]
        melhor_r2 = df_ticker.loc[df_ticker['R2'].idxmax()]
        
        melhores_janelas.append({
            'Ticker': ticker,
            'MAE': melhor_mae['MAE'],
            'MAE_Days': melhor_mae['Janela'],
            'RMSE': melhor_rmse['RMSE'],
            'RMSE_Days': melhor_rmse['Janela'],
            'MSE': melhor_mse['MSE'],
            'MSE_Days': melhor_mse['Janela'],
            'R²': melhor_r2['R2'],
            'R²_Days': melhor_r2['Janela']
        })
    
    return pd.DataFrame(melhores_janelas)

# Criar a tabela combinada
df_melhores_janelas = encontrar_melhor_janela(df)

# Estilizar o DataFrame
styled_df = (df_melhores_janelas
    .style
    .format({
        'MAE': '{:.4f}',
        'RMSE': '{:.4f}',
        'MSE': '{:.4f}',
        'R²': '{:.4f}'
    })
    .background_gradient(
        cmap='RdYlGn_r',
        subset=['MAE', 'RMSE', 'MSE']
    )
    .background_gradient(
        cmap='RdYlGn',
        subset=['R²']
    )
    .set_caption("Comparison of Best Results Across Tickers Micro"))

# Exibir a tabela
styled_df
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Caminho onde estão os arquivos consolidados de métricas
resultados_dir = r"C:\Users\leona\pyhtonscripts\CodigoExperimentos\ExperimentosKfold\resultados_lstm"

# Função para carregar todas as métricas
def carregar_metricas(resultados_dir):
    metricas = []
    for arquivo in os.listdir(resultados_dir):
        if arquivo.endswith("_metrics.csv"):
            filepath = os.path.join(resultados_dir, arquivo)
            try:
                # Tentar carregar o arquivo, ignorando linhas problemáticas
                df = pd.read_csv(filepath, on_bad_lines='skip', engine='python')
                metricas.append(df)
            except Exception as e:
                print(f"Erro ao carregar {arquivo}: {e}")
    return pd.concat(metricas, ignore_index=True)

# Função para encontrar o melhor modelo para cada ticker
def encontrar_melhor_modelo(df, criterio='MAE'):
    melhores_modelos = []
    tickers = df['Ticker'].unique()
    
    for ticker in tickers:
        df_ticker = df[df['Ticker'] == ticker]
        if criterio == 'R2':
            melhor_modelo = df_ticker.loc[df_ticker['R2'].idxmax()]  # Maior R²
        else:
            melhor_modelo = df_ticker.loc[df_ticker[criterio].idxmin()]  # Menor MAE, RMSE ou MSE
        melhores_modelos.append(melhor_modelo)
    
    return pd.DataFrame(melhores_modelos)

# Carregar todas as métricas consolidadas
df_metricas = carregar_metricas(resultados_dir)

# Verificar se o DataFrame está vazio
if df_metricas.empty:
    print("Nenhuma métrica foi carregada. Verifique os arquivos na pasta.")
else:
    # Encontrar o melhor modelo para cada ticker com base no menor MAE
    df_melhores_modelos = encontrar_melhor_modelo(df_metricas, criterio='MAE')

    # Exibir os melhores modelos
    print("Melhores modelos por ticker:")
    print(df_melhores_modelos)

    # Visualizar os resultados em um gráfico
    plt.figure(figsize=(12, 6))
    plt.bar(df_melhores_modelos['Ticker'], df_melhores_modelos['MAE'], color='skyblue', edgecolor='black')
    plt.title('Melhor Modelo por Ticker (Menor MAE)')
    plt.xlabel('Ticker')
    plt.ylabel('MAE')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```
